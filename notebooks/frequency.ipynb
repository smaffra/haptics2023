```python
def intensity(freq, duration, device_frequency, func=np.sin):
    N = device_frequency * duration
    times = np.linspace(0, duration, N) * 2 * 3.14159
    return func(freq * times)
```

```python
plt.plot(times, intensity(freq=3, duration=duration, device_frequency=device_frequency));
```
