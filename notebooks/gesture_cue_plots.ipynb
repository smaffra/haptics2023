## Plotting cue positions

```python
cues = create_gesture_cues(size=10)
idx = 1
plt.figure(figsize=(15,10))
for label, cue in cues.items():
    plt.subplot(3,3, idx)
    cue.plot(duration=1, device_frequency=16000)
    plt.title(label)
    idx+=1
plt.show()
```

## Plotting cue intensities

```python
idx = 1

duration=2
device_frequency = 16000
intensity_frequency=3

N = device_frequency * duration
times = np.linspace(0, duration, N)

plt.figure(figsize=(15,30))
for label, cue in cues.items():
    # sampling cue
    cue_sample = cue.sample(
        duration=duration, 
        device_frequency=device_frequency, 
        intensity_frequency=intensity_frequency)
    # plotting intensities
    for k in range(len(cue_sample)):
        plt.subplot(7,2, idx)
        plt.plot(times, cue_sample[k][:,3])
        plt.title('{}: Control point: {}'.format(label, k))    
        idx+=1
plt.show()
```
